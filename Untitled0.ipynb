Cleaning the data this process is also called Data Preprocessing 

```python
#Clean the text that we got 
import re 
!pip install emot
import pickle
from emot.emo_unicode import UNICODE_EMOJI, EMOJI_UNICODE, EMOTICONS_EMO  
#Create a function to clean the tweets 
def cleanTxT(text):
   text = re.sub(r'@[A-Za-z0-9]+','',text) #Removed @mentions
   text = re.sub(r'#','',text) # removing the '#' symbol 
   text = re.sub(r'\n','',text)
   text = re.sub(r'\?','',text)
   text = re.sub(r'RT[\s]+','',text) #Removing RT(re tweet )
   text = re.sub(r'https?:\/\/\S+' ,'',text) #Remove the hyper link 
  

   return text 

 #Ckeaning the text 
df['tweets'] = df['tweets'].apply(cleanTxT) 


 #Show the clean text 
df
```

To Remove the emoticons or emojis from the text 

```python
import re

def remove_emoji(text):   
    epattern = re.compile("["u"\U0001F600-\U0001F64F" 
                       u'\U0001F300-\U0001F5FF' 
                       u'\U0001F680-\U0001F6FF'
                       u'\U0001F1E0-\U0001F1FF'
                       u'\U00002702-\U000027B0'
                       u'\U000024C2-\U0001F251'
                     "]+", flags = re.UNICODE) 
    
    return epattern.sub(r'',text)  


df['tweets'] = df['tweets'].apply(remove_emoji)

df


   
```

Second step is **Tokenization**

To tokenize meanse to split the string into words

```python
#instantiate tokenizer class 
tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True, 
                                 reduce_len=True )

def tokenize_tweet(text): 

  text = tokenizer.tokenize(text)
  return text 


df['tweets'] = df['tweets'].apply(tokenize_tweet)

df



```

Removing the **stopwords** 

stop words are words that doesnt add any significant meaning

ex  'I' , 'me' 

this is aexample code to explain the working 

```python
nltk.download('stopwords')

# WE WILL IMPORT ENGLISH STOPWORDS LIST FROM NLTK 
stopwords_english = set(stopwords.words('english'))
print(stopwords_english)
```

this funtion will remove the stop words 

```python
nltk.download('stopwords')

# WE WILL IMPORT ENGLISH STOPWORDS LIST FROM NLTK 
stopwords_english = stopwords.words('english')

punctuations = string.punctuation

def remove_stopwords_punctuations(tokenize_tweet): 

  tweet_clean = [] 

  for word in tokenize_tweet:  
    if (word not in stopwords_english and word not in punctuations):
      tweet_clean.append(word)

  return tweet_clean 

df['tweets'] = df['tweets'].apply(remove_stopwords_punctuations) 

df
```

**Stemming**  converts the words to its most general form 

learning -> learn

learned -> learn 

learnt -> learn 

this is a example code to explain the working 

```python
ps = PorterStemmer()

example_words = ["learn","learning","learned"] #list of words to explain stemming

for w in example_words:
   print(ps.stem(w))
```

this function will do the stemming part

```python
stemmer = PorterStemmer()

def get_stem(remove_stopwords_punctuations):

  tweet_stem = []

  for word in remove_stopwords_punctuations:
    stem_word = stemmer.stem(word)
    tweet_stem.append(stem_word)

  return tweet_stem 

df['tweets']=df['tweets'].apply(get_stem)

df
```

Now we will import Subjectivity and Polarity 

```python
# Create a function to get the subjectivity 
from textblob import TextBlob 
def getSubjectivity(text): 
    return TextBlob(text).sentiment.subjectivity  

# Create a function to get the polarity 
def getPolarity(text): 
   return TextBlob(text).sentiment.polarity 


# Create two new coloumn
df['words']=df['tweets'].apply(tokenize_tweet) 
df['Subjectivity']=df['tweets'].apply(getSubjectivity)
df['Polarity']=df['tweets'].apply(getPolarity) 
 


#Show the new Data Frame in new Column 
df 
```

Using the **Naive Bayes Theorem** we will train our model 

```python
# Create a function to compute the negative , neutral and positive analysis 

def getAnalysis(score): 
   if score < 0: 
     return 'Negative' 
   elif score == 0: 
     return 'Neutral' 
   else: 
     return 'Positive' 

df['Analysis'] = df['Polarity'].apply(getAnalysis) 

#show the DataFrame 
df
```
